```python
def MatchWords(s: str, w: List[str]) -> List[int]:

    i_left = 0
    list_matches = []
    mismatch = len(w)
    step = 1
    # step = len(w[0])
    unique_match = False

    d = collections.defaultdict(int)
    for i in w:
        d[i] += 1

    print('w:', w)
    
    print('\n')
    for i_right in range(0, len(s), step):
        
        
        print('i_right:', i_right)
        print('d:', d)
        print('mismatch:', mismatch)
        w_right = s[i_right: i_right + step]
        w_left = s[i_left: i_left + step]
        print('w_right:', w_right)
        print('w_left:', w_left)
        
        
        if (w_right in w) and (d[w_right] > 0):
            
            print(' #1 if w_right in w and d[w_right] > 0:')
            mismatch -= 1
            unique_match = True
    
        
        if w_right in w:
            print(' #2 if w_right in w:')
            d[w_right] -= 1
                
        
            
        print('d:', d)        
        print('mismatch:', mismatch)
        
        # if w_right not in w or d[w_right] < 0:
        #     i_left = i_right
            
        print('d[s[i_left: i_left + step]]',  d[s[i_left: i_left + step]])
        while d[s[i_left: i_left + step]] < 0 and i_left < len(s) - step and i_left < i_right:
            print(' #3 while d[s[i_left: i_left + step]] < 0 and i_left < len(s) - step:')
            d[s[i_left: i_left + step]] += 1
            i_left += step
            
        print('d:', d)
            
            
        print('d[w_right]:', d[w_right])

        if mismatch == 0 and unique_match == True and sum(d.values())==0:
            print(' #4 if mismatch == 0:')
            list_matches.append(i_left)
            unique_match = False
            while s[i_left: i_left + step] not in w:
                i_left += step
            i_left += step
            d[w_left] += 1
            mismatch += 1
            
        # w_left = s[i_left: i_left + step]
        # if (w_left in w) and (d[w_left] < 0):
        #     print(' #4 if (w_left in w) and (d[w_left] < 0):')
        #     i_left += step
        #     d[w_left] += 1
        #     if d[w_left] > 0:
        #         mismatch += 1
        #     if mismatch == 0:
        #         unique_match = True
                
        print('d[s[i_left: i_left + step]]:', d[s[i_left: i_left + step]])
        print('d:', d)
        print('mismatch:', mismatch)
        
        while (s[i_left: i_left + step] not in w) and (i_left < len(s) - step):
            print(' #6 while (s[i_left] not in d) and (i_left < len(s) - step):')
            print(' i_left:', i_left)
            print(' s[i_left]:', s[i_left])
            i_left += step
            
        if mismatch == 0 and unique_match == True and sum(d.values())==0:
            print(' #7 if mismatch == 0:')
            list_matches.append(i_left)
            unique_match = False
        
        print('d[s[i_left: i_left + step]]:', d[s[i_left: i_left + step]])
        print('list_matches:', list_matches)
        print('\n')
            

    return list_matches
```

```python
def find_word_concatenation(s, words):
    if len(words) == 0 or len(words[0]) == 0:
        return []

    word_frequency = collections.defaultdict(int)
    for word in words:
        word_frequency[word] += 1
    
    result_indices = []
    words_count = len(words)
    word_length = len(words[0])
    
    for i in range((len(s) - words_count * word_length)+1):
        words_seen = collections.defaultdict(int)
        for j in range(0, words_count):
            next_word_index = i+j * word_length
            # Get the next word from the string
            word = s[next_word_index: next_word_index + word_length]
            print('word:', word)
            if word not in word_frequency: # Break if we don't need this word
                break
            # Add the word to the 'words_seen' map
            words_seen[word] += 1
            # No need to process further if the word has higher frequency than required
            if words_seen[word] > word_frequency[word]:
                break
            if j + 1 == words_count: # Store index if we have found all the words
                result_indices.append(i)
        print('-------------------')
    return result_indices
```

```python
#  0123456789   
#  0   4   8    12                                     
s="lingmindraboofooowingdingbarrwingmonkeypoundcake"
w=["fooo","barr","wing","ding","wing"]
find_word_concatenation(s, w)
```

```python
s = "barfoofoobarthefoobarman"
w = ["bar","foo","the"]
find_word_concatenation(s, w)
```

```python
# brute force 
# TC: O(n * N * log(N))

def findSubstring(s: str, w: List[str]) -> List[int]:
    if len(w) == 0 or len(s) == 0: return []
    if len(s) < len(''.join(w)): return []

    rst = []
    words_cat = ''.join(sorted(w))
    print('words_cat:', words_cat)
    span = len(w[0])

    for i in range(len(s) - len(words_cat) + 1): 
        s_sub = []
        for j in range(len(w)):
            print(s[i + span * j : i + span * (j+1)])
            s_sub.append(s[i + span * j : i + span * (j+1)])       
        print('s_sub:', s_sub)
        s_sub = ''.join(sorted(s_sub))
        print('s_sub:', s_sub)
        if s_sub == words_cat: rst.append(i)

    return rst 
```
