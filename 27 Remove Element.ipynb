```python
def key_removal(a: List[int], key: int) -> int:
    # if edge_cases != True:
    #     return 'edge cases are not satisfied!'
    
        if a == []:
            return 0

        i = 0  
        while a[i] != key:
            if i+1 == len(a):
                return len(a)
            i += 1

        l = i

        for r in range(l+1, len(a), 1):
            if a[r] != key:
                a[l], a[r] = a[r], a[l]
                l += 1

        return a
        
```
