```python
# brute force 

# Solution 1: bruteforce. O(n^3) time complexity and O(n) space 
# complexity
# time limit exceeded. 230 / 266 test cases passed.
def minWindow(self, s: str, t: str) -> str:
    # helper function to see if the current 
    # window substring has at least equal number 
    # of occurence as the target
    def isCountEqual(target, s_part):
        len_target = len(target.keys())
        i = 0
        for key in target.keys():
            if s_part[key] >= target[key]:
                i += 1
        return i==len_target

    n = len(s) # len of s
    min_win = n+1 # minimum window initialized to something bigger than n
    # target hashtable that contains the chars along with its
    # count of occurence
    target = collections.Counter(t) 
    result = "" # result initialized 
    for i in range(0,n):
        for j in range(i,n):
           # get a dict on the current window substring
           s_part = collections.Counter(s[i:j+1])
           # see if the the current
           # window substring has at least equal number
           # of occurence as the target
           if isCountEqual(target,s_part):
               # if True, then we will check the length of the 
               # current window
                if j-i+1 < min_win: # if less than current min
                    min_win = j-i+1
                    # update the result with minumum window substring
                    result = s[i:j+1] 
    return result
```

```python
# without prints

class Solution:
    def minWindow(self, s: str, p: str) -> str:

        min_len = math.inf
        i_left = 0
        d = collections.defaultdict(int)
        i_left_min = 0
        i_right_min = -1
        
        for i in range(len(p)):
            d[p[i]] -= 1

        for i_right in range(len(s)):

            if s[i_right] in d:
                d[s[i_right]] += 1

            while i <= len(s)-2:
                if s[i_left] in d and d[s[i_left]] > 0:
                    d[s[i_left]] -= 1
                    i_left += 1
                elif s[i_left] not in d:
                    i_left += 1
                else:
                    break
                i += 1   

            if min(d.values()) >= 0 and i_right - i_left < min_len:
                min_len = i_right - i_left
                i_left_min = i_left
                i_right_min = i_right


        return s[i_left_min:i_right_min+1]     
```

```python
def find_substring(pattern, str):
    window_start, matched, substr_start = 0, 0, 0
    min_length = len(str) + 1
    char_frequency = {}
    
    for chr in pattern:
        if chr not in char_frequency:
            char_frequency[chr] = 0
        char_frequency[chr] += 1
    
    # try to extend the range [window_start, window_end]
    for window_end in range(len(str)):
        right_char = str[window_end]
        if right_char in char_frequency:
            char_frequency[right_char] -= 1
            if char_frequency[right_char] >= 0: # Count every matching of a character
                matched += 1
        
        # Shrink the window if we can, finish as soon as we remove a matched character
        while matched == len(pattern):
            if min_length > window_end - window_start + 1:
                min_length = window_end - window_start + 1
                substr_start = window_start
            left_char = str[window_start]
            window_start += 1
            if left_char in char_frequency:
                # Note that we could have redundant matching characters, therefore we'll decrement the
                # matched count only when a useful occurrence of a matched character is going out of the window
                if char_frequency[left_char] == 0:
                    matched -= 1
                char_frequency[left_char] += 1
    
    
    if min_length > len(str):
        return ""
    return str[substr_start: substr_start + min_length]
```

```python
# Solution 2: Sliding window with two pointers and hashtable. 
# O(n) = O(|s|+|t|) time complexity and O(n) = O(|s|+|t|) space 
# complexity
class Solution:
    def minWindow(self, s: str, t: str) -> str:
        # if any of the strinf is empty, nothing to do
        # return empty string
        if not t or not s:
            return ""
        # Otherwise, let's keep an hashtable on target t string
        # that shows which characters how many counts
        dict_t = Counter(t)
        # number of keys in the target t string
        required = len(dict_t.keys())
        # the left and right pointer needed for window sliding
        left, right = 0,0
        # this formed will be incremented while we get a character
        # in out current window that is in t and also it's count
        # at least equal to the t's character count 
        formed = 0
        # containing the current window's character's and their counts
        dict_window = defaultdict(int)
        # will contain the minimum window length and it's left
        # and right index in a tuple 
        result = math.inf, None, None
        n = len(s) # len of s
        # until we cannot proceed further on the string s
        for right in range(0, n):
            char = s[right] # get the char pointed by right
            dict_window[char] += 1 # increment it's count
            # if this char is in t and it's count
            # equal to t's char count, we will say that
            # this char in the current window satifies.
            # at one time it will contain the number that
            # is equal to required
            if char in dict_t and dict_window[char] == dict_t[char]:
                formed += 1
            # at that time, we will check this logic
            # and try to contract window by left increment
            # we will continue to do this as long as the window
            # still satisfies or left get's equal to right
            while left <= right and formed == required:
                char = s[left]
                # before contracting let's record
                # if this the minimum length substring
                if (right-left+1 < result[0]):
                    result = right-left+1, left, right
                # so as you contract means you are kicking out that 
                # character count from the dictionary
                dict_window[char] -= 1
                # if the count for this char becomes less than what
                # is in dict_t, then formed should be also less
                if char in dict_t and dict_window[char] < dict_t[char]:
                    formed -= 1
                
                left += 1 # contracting and get a new window to check
        # return the min window substring by following the result 
        # tuple
        return s[result[1]: result[2] + 1] if result[0] != math.inf else ""
```

```python
def minWindow(s, t):
    need = collections.Counter(t)            #hash table to store char frequency
    missing = len(t)                         #total number of chars we care
    start, end = 0, 0
    i = 0
    for j, char in enumerate(s, 1):          #index j from 1
        if need[char] > 0:
            missing -= 1
        need[char] -= 1
        if missing == 0:                     #match all chars
            while i < j and need[s[i]] < 0:  #remove chars to find the real start
                need[s[i]] += 1
                i += 1
            need[s[i]] += 1                  #make sure the first appearing char satisfies need[char]>0
            missing += 1                     #we missed this first char, so add missing by 1
            if end == 0 or j-i < end-start:  #update window
                start, end = i, j
            i += 1                           #update i to start+1 for next window
    return s[start:end]
```
