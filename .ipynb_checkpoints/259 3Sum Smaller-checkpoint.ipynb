```python
import math
from typing import List
import collections
```

```python
def triplet_sum_less_then_target(a: List[int], t: int) -> int:

    # if edge_cases(a) == False: return [[]]:
    a.sort()
    count = 0
    # results = []

    for i in range(0, len(a)-2, 1):
        # indexes could be not unique, so we don't need it
        # if i < len(a)-2 and i > 0 and a[i] == a[i-1]:
        #     print( 'continue')        
        #     continue

        # Since the array is sorted, the lower bound of 3sum will be 3*nums[i]
        if 3*a[i] >= t:
            break

        l, r = i+1, len(a)-1

        while l < r:

            s = a[i] + a[l] + a[r] 

            if s >= t:
                r -= 1
                    
            else:
                
                # if we want to see triples
                # for j in range(r, l, -1):
                #     results.append([a[i], a[l], a[j]])
                #     print(results)
                
                # if (i,j,k) works, then (i,j,k), (i,j,k-1),..., 
                # (i,j,j+1) all work, totally (k-j) triplets
                count += r-l
                l += 1
                
            # print('\n')


    return count
```

```python
 #  [-4, -2, -2, -2, 0, 1, 2, 2, 2, 3, 3, 4, 4, 6, 6]
a = [3,2,-2,6,2,-2,6,-2,-4,2,3,0,4,4,1]
t = 3
triplet_sum_less_then_target(a, t)
```

```python
 #   -1, 0, 2, 3
a = [-1, 0, 2, 3]
t = 3
triplet_sum_less_then_target(a, t)
# [-1, 0, 3], [-1, 0, 2]
```

```python
#    -1, 1, 2, 3, 4
a = [-1, 4, 2, 1, 3]
t = 5
triplet_sum_less_then_target(a, t)
# [-1, 1, 4], [-1, 1, 3], [-1, 1, 2], [-1, 2, 3]
```
