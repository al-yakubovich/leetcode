```python
# N log N --> sorted(s1)
# M log M --> sorted(sub)
# TC: O( N log N + N (M log M + M) ) = {M ~ N} = O( N log N + N (M log N + N) ) ~ N^2 
# TC: O( N log N + N (M log M + M) ) = {M << N} = N log N

def checkInclusion(s1: str, s2: str) -> bool:
    s1_sorted = ''.join(sorted(s1))
    for i in range(0, len(s2) - len(s1) + 1):
        print('i:', i)
        sub = s2[i: i + len(s1)]
        print('sub:', sub)
        sub_sorted = ''.join(sorted(sub))

        if sub_sorted == s1_sorted:
            return True
    return False
```

```python
checkInclusion('abc', 'tacbg')
```

```python
def permute(a, l, length):
    if l==length: 
        print(''.join(a) )
    else:
        for j in range(l,length): 
            print('l', l)
            print('j', j)
            print('a[l], a[j]',a[l], a[j])
            a[l], a[j] = a[j], a[l] 
            permute(a, l+1, length) 
            a[l], a[j] = a[j], a[l]  
            print('=======================')

            
string = "ABC"
permute(list(string), 0, len(string))
```

```python
import collections

def pattern_detector(p, s):
    d_pattern = collections.defaultdict(int)
    for i in p:
        d_pattern[i] += 1
        
    d_current = collections.defaultdict(int)
    
    i_left = 0
    pattern_detected = False
    
    for i_right in range(len(s)):
        
        d_current[s[i_right]] += 1
            
        if d_current == d_pattern:
            pattern_detected = True
            break
        
        if i_right - i_left + 1 == len(p):
        
            d_current[s[i_left]] -= 1
            
            if d_current[s[i_left]] == 0:
                del d_current[s[i_left]]
                        
            i_left += 1                
        
        
    return pattern_detected
```

```python
s1 = "ABC"
s2 = "TACBG"
pattern_detector(s1, s2)
```

```python
s1 = "ABC"
s2 = "TACCBG"
pattern_detector(s1, s2)
```

```python
def find_permutation(pattern, str):
    window_start, matched = 0, 0
    char_frequency = {}
    for chr in pattern:
        if chr not in char_frequency:
            char_frequency[chr] = 0
        char_frequency[chr] += 1
        
    # our goal is to match all the characters from the 'char_frequency' with the current window
    # try to extend the range [window_start, window_end]
    for window_end in range(len(str)):
        right_char = str[window_end]
        if right_char in char_frequency:
            # decrement the frequency of matched character
            char_frequency[right_char] -= 1
            if char_frequency[right_char] == 0:
                matched += 1
        if matched == len(char_frequency):
            return True
        # shrink the window by one character
        if window_end >= len(pattern) - 1:
            left_char = str[window_start]
            window_start += 1
            if left_char in char_frequency:
                if char_frequency[left_char] == 0:
                    matched -= 1
                char_frequency[left_char] += 1
    return False
```

```python
s1 = "ABC"
s2 = "TACBG"
find_permutation(s1, s2)
```

```python
s1 = "ABC"
s2 = "TACCBG"
find_permutation(s1, s2)
```

```python
def checkInclusion(s1, s2):
    A = [ord(x) - ord('a') for x in s1]
    B = [ord(x) - ord('a') for x in s2]
    
    target = [0] * 26
    for x in A:
        target[x] += 1
    
    window = [0] * 26
    for i, x in enumerate(B):
        window[x] += 1
        if i >= len(A):
            window[B[i - len(A)]] -= 1
        if window == target:
            return True
    return False
```
